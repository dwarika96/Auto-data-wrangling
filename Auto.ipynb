```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
#read data
df=pd.read_csv('//content//sample_data//auto.csv')
```

```python
df.head()
```

```python
#conver '?' values into nan value
df.replace('?',np.nan,inplace=True)
```

```python
df.head()
```

```python
#give column name
df.columns=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.head()
```

```python
#find the null values
df.isnull().sum()
```

```python
#find data type of the data
df.info()
```

```python
#convert required columns from object to float
df['normalized-losses']=df['normalized-losses'].astype('float')
df['bore']=df['bore'].astype('float')
df['stroke']=df['stroke'].astype('float')
df['horsepower']=df['horsepower'].astype('float')
df['peak-rpm']=df['peak-rpm'].astype('float')
df['price']=df['price'].astype('float')
```

```python
df.isnull().sum()
```

```python
x=df['normalized-losses'].mean()
df['normalized-losses'].fillna(x,inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df['bore'].median()
df['bore'].fillna(x,inplace=True)
```

```python
x=df['stroke'].median()
df['stroke'].fillna(x,inplace=True)
```

```python
x=df['horsepower'].mode()[0]
df['horsepower'].fillna(x,inplace=True)
```

```python
x=df['peak-rpm'].mode()[0]
df['peak-rpm'].fillna(x,inplace=True)
```

```python
x=df['price'].mode()[0]
df['price'].fillna(x,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#transform mpg to L/100km in the column of "highway-mpg", and change the name of column to "highway-L/100km".
df['highway-L/100km']=df['highway-mpg'].max()/df['highway-mpg']
```

```python
df.head()
```

```python
#Normalize those variables so their value ranges from 0 to 1
df['Normal_length']=df['length']/df['length'].max()
df['Normal_width']=df['width']/df['width'].max()
df['Normal_height']=df['height']/df['height'].max()
```

```python
df.head()
```

```python
#Binning Horse power
bin=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
#set group name
name=['Low','Medium','High']
```

```python
#apply bin,name and create new column
df['Horsepower_binning']=pd.cut(df['horsepower'],bin,labels=name,include_lowest=True)
```

```python
df[['horsepower','Horsepower_binning']].head(20)
```

```python
df['Horsepower_binning'].value_counts()
```

```python
df.head()
```

```python
plt.bar(name,df['Horsepower_binning'].value_counts())
plt.xlabel('Horse power')
plt.ylabel('count')
plt.title('Horsepower bins')
plt.show()
```

```python
plt.plot(name,df['Horsepower_binning'].value_counts())
plt.show()
```
